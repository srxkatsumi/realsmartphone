# Previsão de Adoção de Tecnologia 5G # 
# 5G Technology Adoption Forecast #


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Gerando dados de exemplo
np.random.seed(0)
anos = np.arange(2010, 2021)
taxa_adoção_5G = np.array([0.1, 0.2, 0.4, 0.6, 0.7, 0.8, 0.85, 0.88, 0.9, 0.92, 0.95])
dados = np.column_stack((anos, taxa_adoção_5G))
```

```python
# Separando os dados em treinamento e teste
X = dados[:, 0].reshape(-1, 1)
y = dados[:, 1]
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X_treinamento, y_treinamento)
```

```python
# Fazendo previsões
previsões = modelo.predict(X_teste)
```

```python
# Visualização dos resultados
plt.scatter(X, y, label='Dados reais')
plt.plot(X_teste, previsões, color='red', linewidth=2, label='Previsões')
plt.xlabel('Ano')
plt.ylabel('Taxa de Adoção de 5G')
plt.legend()
plt.title('Previsão da Adoção de 5G ao longo do Tempo')
plt.show()
```
